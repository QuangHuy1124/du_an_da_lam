```python
df['age'] = df['age'].astype(int)
```

```python
# Tạo function đếm số dòng null và tỷ lệ % của từng column :
def calc_null_rate(df):
    newdata = df.isnull().sum().to_frame('null_count')
    newdata[['null_rate']] = newdata[['null_count']] / len(df)
    return newdata.sort_values(by=['null_rate'], ascending=False)
```

# 3. Phân tích Dữ liệu Thăm dò (EDA)

Tỷ lệ bệnh nhân bị đột quỵ trong bộ dữ liệu

```python
stroke_rate = df["stroke"].mean() * 100
stroke_count = df["stroke"].sum()
total = len(df)

print(f"Số ca đột quỵ: {stroke_count}/{total} ({stroke_rate:.2f}%)")
```

Tỷ lệ bệnh nhân bị đột quỵ khoảng 5% và không bị đột quỵ là 95% cho thấy bộ dữ liệu có sự mất cân bằng nghiêm trọng.

Bộ dữ liệu mất cân bằng nghiêm trọng như vậy sẽ dẫn đến:
   
- Mô hình dễ thiên lệch về lớp chiếm đa số (không đột quỵ), từ đó có thể đạt độ chính xác (accuracy) cao nhưng không phát hiện được nhiều ca đột quỵ thật sự.

```python
# Tính tuổi trung bình theo tình trạng đột quỵ
mean_age = df.groupby("stroke")["age"].mean().reset_index()

plt.figure(figsize=(6,5))
ax = sns.barplot(x="stroke", y="age", data=mean_age, palette="Set2")

plt.title("Tuổi trung bình theo tình trạng đột quỵ")
plt.xlabel("Tình trạng đột quỵ (0 = Không, 1 = Có)")
plt.ylabel("Tuổi trung bình")

# Điền số vào cột
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}',
                (p.get_x() + p.get_width()/2., height),
                ha='center', va='bottom', fontsize=11, color='black')

plt.show()
```

Tình trạng đột quỵ có tuổi trung bình cao hơn rõ rệt. Cho thấy tuổi tác là một yếu tố quan trọng dẫn đết đột quỵ.

```python
# Tính số lượng và tỷ lệ phần trăm theo bệnh tim và đột quỵ
heart_disease_stroke_counts = df.groupby(['heart_disease', 'stroke']).size().unstack(fill_value=0)
heart_disease_stroke_counts['Total'] = heart_disease_stroke_counts.sum(axis=1)

# Chuẩn bị dữ liệu cho biểu đồ tròn
labels = ['Không Bệnh Tim - Không Đột Quỵ', 'Không Bệnh Tim - Đột Quỵ',
          'Có Bệnh Tim - Không Đột Quỵ', 'Có Bệnh Tim - Đột Quỵ']
sizes = [
    heart_disease_stroke_counts.loc[0, 0] if 0 in heart_disease_stroke_counts.index else 0,
    heart_disease_stroke_counts.loc[0, 1] if 0 in heart_disease_stroke_counts.index else 0,
    heart_disease_stroke_counts.loc[1, 0] if 1 in heart_disease_stroke_counts.index else 0,
    heart_disease_stroke_counts.loc[1, 1] if 1 in heart_disease_stroke_counts.index else 0
]
colors = ['#66CCCC', '#FF9999', '#66CC99', '#FF6666']  # Gradient màu: xanh nhạt, đỏ nhạt, xanh lá nhạt, đỏ đậm
explode = (0.05, 0.05, 0.05, 0.05)  # Tách nhẹ các phần để nổi bật

# Thiết lập style
plt.rcParams['font.family'] = 'Arial'
fig, ax = plt.subplots(figsize=(8, 8))

# Vẽ biểu đồ tròn
wedges, texts, autotexts = ax.pie(sizes, explode=explode, colors=colors,
                                 autopct='%1.1f%%', shadow=True, startangle=90,
                                 textprops={'fontsize': 8, 'fontweight': 'bold', 'color': 'black'},
                                 wedgeprops={'edgecolor': 'black', 'linewidth': 1.2})

# Tùy chỉnh tiêu đề
ax.set_title('Tỷ Lệ Đột Quỵ Theo Bệnh Tim', fontsize=16, fontweight='bold', pad=20)

# Đảm bảo biểu đồ tròn không bị méo
ax.axis('equal')

# Thêm chú thích
plt.legend(wedges, labels, title="Bệnh Tim & Đột Quỵ", loc="best", fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Tính số lượng và tỷ lệ phần trăm theo giới tính
gender_counts = stroke_data['gender'].value_counts()
total_stroke1 = len(stroke_data)
gender_percentages = (gender_counts / total_stroke1 * 100).round(1)

# Thiết lập style
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'Arial'
fig, ax = plt.subplots(figsize=(6, 4))  # Kích thước nhỏ gọn

# Vẽ biểu đồ cột với màu sắc khác nhau cho từng cột
colors = {'Male': '#1E90FF', 'Female': '#FF69B4'}  # Xanh dương cho Male, hồng cho Female
bars = []
for i, gender in enumerate(gender_counts.index):
    height = gender_counts[gender]
    bar = ax.bar(gender, height, color=colors[gender],
                 edgecolor='black', linewidth=1.5, alpha=0.9)
    bars.append(bar)

# Thêm tỷ lệ phần trăm trên cột (sửa lỗi bằng cách lặp qua các Patch)
for i, bar_container in enumerate(bars):
    for bar in bar_container:  # Lặp qua từng Patch trong BarContainer
        height = bar.get_height()
        if height > 0:
            percentage = gender_percentages.iloc[i]
            ax.annotate(f'{percentage}%',
                        (bar.get_x() + bar.get_width() / 2., height / 2),
                        ha='center', va='center', fontsize=10, fontweight='bold', color='white')

# Tùy chỉnh tiêu đề và nhãn
ax.set_title('Tỷ Lệ Đột Quỵ Theo Giới Tính (Chỉ Stroke=1)', fontsize=12, fontweight='bold', pad=10)
ax.set_xlabel('Giới Tính', fontsize=10, fontweight='bold')
ax.set_ylabel('Số Lượng', fontsize=10, fontweight='bold')

# Loại bỏ legend vì chỉ có stroke=1, màu sắc đã phân biệt
# ax.get_legend().remove()

# Tùy chỉnh lưới và viền
ax.grid(True, linestyle='--', alpha=0.6, axis='y')
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

# 5. Xây dựng mô hình dự đoán

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report, f1_score
```

```python
# Xử lý giá trị NaN (điền bằng giá trị trung bình cho biến số liên tục)
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
df['smoking_status'] = df['smoking_status'].fillna('Unknown')

# Mã hóa biến phân loại
label_encoders = {}
for column in ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))  # Chuyển đổi thành số
    label_encoders[column] = le

# Chia đặc trưng (features) và nhãn (target)
X = df[['age', 'avg_glucose_level', 'bmi', 'hypertension', 'heart_disease',
        'gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']]
y = df['stroke']

# Chuẩn hóa các đặc trưng
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Tạo và huấn luyện mô hình Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test)

# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
print(f"Độ chính xác: {accuracy:.2f}")
print("\nBáo cáo phân loại:")
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
# Hiển thị các tham số của mô hình
print("\nCác tham số của mô hình:")
print(model.get_params())
```

```python
df['smoking_status'] = df['smoking_status'].fillna('Unknown')
label_encoders = {}
for column in ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

# Đặc trưng và nhãn
X = df[['age', 'avg_glucose_level', 'bmi', 'hypertension', 'heart_disease', 'gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']]
y = df['stroke']

# Chuẩn hóa
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# SMOTE cho mất cân bằng
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_scaled, y)

# Chia dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Huấn luyện mô hình Logistic Regression
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)

# Dự đoán và đánh giá
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred)
print(f"Logistic Regression - F1-Score: {f1:.2f}")
print("Logistic Regression - Báo cáo phân loại:")
print(classification_report(y_test, y_pred))
```
